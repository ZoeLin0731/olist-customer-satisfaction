### Step4: Convert datetime columns 

```python
date_cols_orders = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols_orders:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")
```

```python
# Datetime columns in reviews table
date_cols_reviews = [
    "review_creation_date",
    "review_answer_timestamp"
]

for col in date_cols_reviews:
    reviews[col] = pd.to_datetime(reviews[col], errors="coerce")
```

```python
orders[date_cols_orders].dtypes
reviews[date_cols_reviews].dtypes
```

### Step5: Define FIRST business metrics
**Delivery delay**

```python
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"]
    - orders["order_estimated_delivery_date"]
).dt.days
```

**Customer dissatisfaction flag**

```python
reviews["is_dissatisfied"] = reviews["review_score"] <= 2
```

### Step6: Merge tables

```python
orders_reviews = orders.merge(
    reviews[["order_id", "review_score", "is_dissatisfied"]],
    on="order_id",
    how="left"
)
```

```python
orders_reviews[["delivery_delay_days", "review_score"]].describe()
```

```python
orders_reviews.groupby("is_dissatisfied")["delivery_delay_days"].agg(
    ["count", "mean", "median"]
)
```

### Step7: Initial EDA visualizations
**Distribution of delivery delays**

```python
plt.figure(figsize=(8,4))
sns.histplot(
    orders_reviews["delivery_delay_days"],
    bins=50,
    kde=True
)
plt.title("Distribution of Delivery Delay (Days)")
plt.xlabel("Delivery Delay (Days)")
plt.show()
```

**Relationship: delay vs satisfaction**

```python
plt.figure(figsize=(8,4))
sns.boxplot(
    x="is_dissatisfied",
    y="delivery_delay_days",
    data=orders_reviews
)
plt.title("Delivery Delay vs Customer Dissatisfaction")
plt.xlabel("Is Dissatisfied (<=2 stars)")
plt.ylabel("Delivery Delay (Days)")
plt.show()
```

### Initial Observations
- Orders delivered later than estimated are significantly more likely to receive 2-star reviews
- Median delivery delay for dissatisfied customers is higher than for satisfied customers
- However, not all dissatisfaction is explained by delivery delay, indicating additional drivers (product quality, seller behavior)
- Delivery performance is a strong but not exclusive contributor to customer satisfaction
